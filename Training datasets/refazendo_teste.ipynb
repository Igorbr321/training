```python
import pandas as pd
```

```python
sheets = ['Contratos', 'Pagamentos', 'Estrutura comercial closers', 'Estrutura comercial', 'Regras de comissionamento']

dfs = {}

for sheet in sheets:
    dfs[sheet] = pd.read_excel("W1 Holdings - Teste técnico - Analista de Dados.xlsx", sheet_name=sheet)
```

```python
df_comissionamento = dfs['Regras de comissionamento']

df_comissionamento
```

```python
df_pagamentos = dfs['Pagamentos']

df_pagamentos
```

```python
df_results = df_pagamentos.assign(
    consultor = df_pagamentos['Valor'] * 0.10,
    lider = df_pagamentos['Valor'] * 0.05,
    closer = df_pagamentos['Valor'] * 0.20,
    lider_closer = df_pagamentos['Valor'] * 0.15
)

```

```python
df_results
```

```python
df_contratos = dfs['Contratos']

df_results = df_results.merge(df_contratos, on='ID de negócio', how='left')
```

```python
df_results
```

```python
df_results = df_results[['ID de negócio', 'Data de pagamento', 'Valor', 'consultor', 'lider', 'closer', 'lider_closer','ID Consultor', 'ID Closer']]

df_results.head(10)
```

```python
df_results.dtypes
```

```python
cols_to_convert = ['ID Consultor', 'ID Closer']

df_results[cols_to_convert] = df_results[cols_to_convert].fillna(0).astype(int)

```
